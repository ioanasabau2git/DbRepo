```python
from pathlib import Path
out_dir = Path('/Volumes/openaq/bronze_dev/locations')
out_dir.mkdir(parents=True, exist_ok=True)
fname = out_dir / f'locations_{datetime.datetime.now().strftime("%Y-%m-%d")}.json'
with open(fname, 'w') as f:
    f.write(json.dumps(data))
print('wrote', fname)
```
